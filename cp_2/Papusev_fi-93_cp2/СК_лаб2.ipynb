```python
def shifr (key):
    f = open('Text')
    i=0
    Shifrotext=str()
    for char in f.read():
      if ord(char)>=1072 and ord(char)<=1072+31:
        Shifrotext=Shifrotext+str(chr((ord(char)-1072+ord(key[i])-1072)%32+1072))
        i=i+1
        if i>=len(key)-1:
          i=0
    return Shifrotext
    f.close()
```

```python
def Indeks(Y):
  I=0
  for i in range(32):
    I=I+count_liters(Y)[i]*(count_liters(Y)[i]-1)
  I=I/ALL(Y)/(ALL(Y)-1)
  return I
```

```python
def count_liters(Y):
  count=[ 0 for i in range(32)]
  for char in Y:
    if ord(char)>=1072 and ord(char)<=1072+31:
      count[ord(char)-1072]=count[ord(char)-1072]+1
  return count
```

```python
def ALL(Y):
   All=0
   for char in Y:
     if ord(char)>=1072 and ord(char)<=1072+31:
        All=All+1
   return All
```
